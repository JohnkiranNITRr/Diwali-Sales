```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

# Data Cleaning

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace= True)
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Amount'] = (df['Amount'] ).astype('int')
```

```python
df['Amount'] .dtypes
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status' : 'Shaadi'} )
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

```python
df.columns
```

```python
sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x= 'Gender', y = 'Amount', data = sales_gen)
```

```python
ax1 = sns.barplot(x= 'Gender', y = 'Amount', data = sales_gen)
# Labels:
for bars in ax1.containers:
    ax1.bar_label(bars)
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x= 'Age Group', y = 'Amount', data = sales_age)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ax = sns.barplot(x= 'Age Group', y = 'Amount', data = sales_age)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sales = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
plt.figure(figsize = (18,5))
sns.barplot(x = 'State', y= 'Amount', data = state_sales)
```

```python
state_orders = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
plt.figure(figsize = (18, 6))
sns.barplot(x = 'State', y= 'Orders', data = state_orders)
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'Shaadi', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
shaadi_orders = df.groupby(['Shaadi', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
plt.figure(figsize = (5,5))
sns.barplot(x= 'Shaadi', y= 'Amount', data = shaadi_orders, hue = 'Gender')
```

```python
df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
occu_sales = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
plt.figure(figsize = (18,5))
sns.barplot(x  = 'Occupation', y  = 'Amount', data = occu_sales)
```

```python
plt.figure(figsize = (18, 5))
ax = sns.countplot(x= 'Occupation', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
plt.figure(figsize = (25, 5))
ax = sns.countplot(x= 'Product_Category', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
```

```python
category_sales = df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
plt.figure(figsize = (35,5))
sns.barplot(x= 'Product_Category', y= 'Orders', data = category_sales)
```

```python
category_sales = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
plt.figure(figsize = (35,5))
sns.barplot(x= 'Product_Category', y= 'Amount', data = category_sales)
```

```python
category_sales = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
plt.figure(figsize = (35,5))
sns.barplot(x= 'Product_ID', y= 'Orders', data = category_sales)
```
